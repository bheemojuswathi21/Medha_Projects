**1.Print customer Id, customer name and average account_balance maintained by each customer for all of his/her accounts in the bank.(8 Rows)**

```python
# SQL query to get customer_id, customer_name, and average balance amount
query = """
SELECT
    c.customer_id,
    c.customer_name,
    AVG(a.Balance_amount) AS avg_balance
FROM Bank_Customer c
LEFT JOIN Bank_Account_Detail2 a ON c.customer_id = a.customer_id
GROUP BY c.customer_id, c.customer_name
LIMIT 8;  -- Show only 8 rows
"""

# Execute the query
cursor.execute(query)

# Fetch column names
columns = [desc[0] for desc in cursor.description]

# Fetch all results
results = cursor.fetchall()

# Print table headers
print("{:<12} {:<20} {:<15}".format(*columns))
print("-" * 50)

# Print each row in a properly formatted way
for row in results:
    print("{:<12} {:<20} {:<15.2f}".format(*row))  # Format avg_balance to 2 decimal places
```

**2) Print customer_id , account_number and balance_amount , condition that if balance_amount is nil then assign transaction_amount for account_type = "Credit Card"(4 Rows).**

```python
query = """
SELECT
    a.customer_id,
    a.Account_number,
    COALESCE(a.Balance_amount, t.Transaction_amount) AS Balance_amount
FROM Bank_Account_Detail2 a
LEFT JOIN BANK_ACCOUNT_Transaction1 t ON a.Account_Number = t.Account_Number
WHERE a.Account_type = 'Credit Card'
LIMIT 4;  -- Show only 4 rows
"""
result = pd.read_sql_query(query, conn)
print(tabulate(result, headers='keys', tablefmt='psql'))
```

```python
# 3) Print account_number and balance_amount , transaction_amount,Transaction_Date from Bank_Account_Details and bank_account_transaction
    #for all the transactions occurred during march,2020 and april, 2020(12 Rows)
query = """
SELECT
    a.Account_Number,
    a.Balance_amount,
    t.Transaction_amount,
    t.Transaction_Date
  from Ban_Account_Details a
  INNER Join BANK_ACCOUNT_Transaction1 t on a.Account_Number = a.Account_Number
  WHERE t.Transaction_Date BETWEEN '2020-03-01' AND '2020-04-30'
  limit 12;
"""

result = pd.read_sql_query(query, conn)
print(tabulate(result, headers='keys', tablefmt ='psql'))
```

```python
# 4) Print all of the customer id, account number,  balance_amount, transaction_amount , Transaction_Date
# from bank_customer, Bank_Account_Details and bank_account_transaction tables where excluding all of their transactions in march, 2020  month (22 Rows)

query = """
SELECT
    c.customer_id,
    a.Account_Number,
    a.Balance_amount,
    t.Transaction_amount,
    t.Transaction_Date
FROM Bank_Customer c
INNER JOIN Bank_Account_Detail2 a
    ON c.customer_id = a.customer_id
INNER JOIN Bank_Account_Transaction1 t
    ON a.Account_Number = t.Account_Number
WHERE
    NOT (t.Transaction_Date BETWEEN '2020-03-01' AND '2020-03-31')
LIMIT 22;
"""

result = pd.read_sql_query(query, conn)
print(tabulate(result, headers='keys', tablefmt ='psql'))
```

```python
# 5) Print only the customer id, account_number, balance_amount,transaction_amount ,transaction_date who did transactions during
#the first quarter. Do not display the accounts if they have not done any transactions in the first quarter.(16 Rows)

query = """
SELECT
    c.Customer_id,
    a.Account_Number,
    a.Balance_amount,
    t.Transaction_amount,
    t.Transaction_Date
FROM bank_customer c
INNER JOIN Bank_Account_Detail2 a
    ON c.Customer_id = a.Customer_id
INNER JOIN Bank_Account_Transaction1 t
    ON a.Account_Number = t.Account_Number
WHERE t.Transaction_Date BETWEEN '2020-01-01' AND '2020-03-31'
LIMIT 16;
"""


result = pd.read_sql_query(query, conn)
print(tabulate(result, headers='keys', tablefmt ='psql'))
```

```python
# 6) Print account_number, Event and Customer_message from BANK_CUSTOMER_MESSAGES and Bank_Account_Details
#to display an Adhoc" Event for all customers who have  SAVINGS" account_type account.(8 Rows)
query = """
SELECT
    a.Account_Number,
    m.Event,
    m.Customer_message
FROM BANK_CUSTOMER_MESSAGE m
CROSS JOIN Bank_Account_Detail2 a
WHERE m.Event = 'Adhoc' AND a.Account_type = 'SAVINGS'
LIMIT 8;
"""

result = pd.read_sql_query(query, conn)
print(tabulate(result, headers='keys', tablefmt ='psql'))
```

```python

# 7) Print all Customer_id, Account_Number, Account_type, and display deducted balance_amount by  subtracting only
#negative transaction_amounts for Relationship_type ="P" ( P - means  Primary , S - means Secondary ) .(27 Rows)

query = """
SELECT
    a.customer_id,
    a.Account_Number,
    a.Account_type,
    SUM(CASE WHEN t.Transaction_amount < 0 THEN t.Transaction_amount ELSE 0 END) AS deducted_balance_amount
FROM Bank_account_Detail2 a
JOIN Bank_Account_Transaction1 t
    ON a.Account_Number = t.Account_Number
WHERE a.relationship = 'P'
GROUP BY a.customer_id, a.Account_Number, a.Account_type
LIMIT 27;
"""

result = pd.read_sql_query(query, conn)
print(tabulate(result, headers='keys', tablefmt ='psql'))
```

```python

# 10) Display all saving account holders have Add-on Credit Cards" and Credit cards" (3 Rows)
query = """
SELECT
    c.Customer_id,
    c.Customer_name,
    GROUP_CONCAT(a.Account_Number) AS Account_Numbers
FROM Bank_Customer c
JOIN Bank_Account_Detail2 a
      ON c.Customer_id = a.Customer_id
WHERE a.Account_type IN ('SAVINGS', 'Credit Card', 'Add-on Credit Card')
GROUP BY c.Customer_id, c.Customer_name
"""
result = pd.read_sql_query(query, conn)
print(tabulate(result, headers='keys', tablefmt='grid'))
```

```python
# 11)  Display  records of SAVINGS account linked with Credit cards" account_type and its credit
# aggregate sum of transaction amount. (1 Row)
# Ref: Check linking relationship in bank_transaction_relationship_details.
        # Check transaction_amount in bank_account_transaction.


query = """
SELECT
    a.Account_Number,
    SUM(t.Transaction_amount) AS TOTAL_SUM
FROM Bank_Account_Detail2 a
JOIN Bank_Account_Transaction1 t
    ON a.Account_Number = t.Account_Number
JOIN BANK_ACCOUNT_Relationship_Detail r
    ON a.Account_Number = r.Linking_Account_Number
WHERE a.Account_type = 'SAVINGS'
  AND r.Account_type = 'Credit Card'
GROUP BY a.Account_Number;
"""

result11 = pd.read_sql_query(query, conn)
print(tabulate(result11, headers="keys", tablefmt="grid"))
```
